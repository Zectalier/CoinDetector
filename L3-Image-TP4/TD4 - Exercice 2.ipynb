```python
import numpy as np
import skimage
from matplotlib import pyplot as plt
```

```python
img = skimage.io.imread('document__.jpeg', as_gray=True)


img = np.where(img < np.mean(img), 1, 0)

plt.imshow(img, cmap='gray')
plt.show()
```

```python
mean_per_line = np.mean(img, axis=1)
```

```python
print(len(np.where(mean_per_line == 1)[0]))

plt.plot(mean_per_line, np.arange(0, len(mean_per_line)))
plt.show()
```

```python
nb_lines = 0
seuil = 0.1
lineFound = False

for mean in mean_per_line:
    if mean > seuil and not lineFound:
        lineFound = True
        nb_lines += 1
    if lineFound and mean < seuil:
        lineFound = False
        
print(nb_lines)
```
