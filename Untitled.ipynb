```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
import time
```

```python
elementStruct = np.ones((9,9))
centre = [4,4]

srcimg = cv2.imread("imagetest.jpg",cv2.IMREAD_GRAYSCALE)

src = np.zeros((srcimg.shape[0],srcimg.shape[1]))
src[srcimg<50] = 0
src[srcimg>50] = 1
```

```python
plt.imshow(src, cmap='gray')
plt.plot()
```

```python
def erode_img(img, elementStruct, center):
    e1 = int(elementStruct.shape[0]/2)
    e2 = int(elementStruct.shape[1]/2)
    print(str(e1)+str(e2))
    dst = np.zeros((img.shape[0],img.shape[1]))
    for i in range(e1,img.shape[0]-e1):
        for j in range(e2,img.shape[1]-e2):
            array = []
            for k in range(elementStruct.shape[0]):
                line = []
                for l in range(elementStruct.shape[1]):
                    line.append(img[i-e1+k][j-e2+l])
                array.append(line)
            nparray = np.asmatrix(array)
            if erode(nparray,elementStruct):
                x = i-e1+center[0]
                y = j-e2+center[1]
                dst[x,y] = 1
                print("x:"+str(x)+" y:"+str(y)+" "+str(dst[x,y]))
    return dst
```

```python
def erode(array, elementStruct):
    toErode = True
    for i in range(array.shape[0]):
        for j in range(array.shape[1]):
            if array[i,j]!=elementStruct[i,j]:
                if elementStruct[i,j]==1:
                    toErode = False
    
    return toErode
```

```python
dst = erode_img(src, elementStruct, centre)
```

```python
plt.imshow(dst, cmap='gray')
plt.plot()
```
