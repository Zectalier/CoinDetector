```python
import skimage
import numpy as np
from matplotlib import pyplot as plt
from os import listdir
from os.path import isfile, join
import cv2
```

```python
Y_PATH = './data/train_labels/'
Y_HAT_PATH = './Binarisation/prediction_labels_2/'
```

```python
labels = [f for f in listdir(Y_PATH) if isfile(join(Y_PATH, f))]
predictions = [f for f in listdir(Y_HAT_PATH) if isfile(join(Y_HAT_PATH, f))]
print(str(len(labels)) + ' labeled images found in folder: ' + Y_PATH)
print(str(len(predictions)) + ' predicted images found in folder: ' + Y_PATH)
```

```python
processed = 0
total_accuracy = 0

for pred in predictions:
    y = cv2.imread(Y_PATH + pred.split('.')[0] + '.png', 0)
    y_hat = cv2.imread(Y_HAT_PATH + pred, 0)
    y_hat = np.where(y_hat == 0, 1, 0)
    
    try:
        y_and_y_hat = y * y_hat
        y_or_y_hat = y + y_hat

        """y_TP = len(np.where(y_and_y_hat == 0)[0])
        y_FP = len(np.where(y - y_hat == 0)[0])
        y_FN = len(np.where(y_hat - y == 0)[0])
        y_TN = len(np.where(y_hat - y == 1)[0])

        print('y_TP = ' + str(y_TP))
        print('y_FP = ' + str(y_FP))
        print('y_FN = ' + str(y_FN))
        print('y_TP = ' + str(y_TP))"""

        accuracy = len(np.where(y_or_y_hat == 0)[0]) / len(np.where(y_and_y_hat == 0)[0])
        total_accuracy += accuracy
        processed += 1
    except:
        print('Predicted label not in labels.')

print('Total Accuracy: ' + str((total_accuracy / processed) * 100) + ' %.')
```

```python
processed = 0
total_accuracy = 0

for pred in predictions:
    y = cv2.imread(Y_PATH + pred.split('.')[0] + '.png', 0)
    y = np.where(y == 255, 1, 0)
    y_hat = cv2.imread(Y_HAT_PATH + pred, 0)
    y_hat = np.where(y_hat > 150, 0, 1)
    
    try:
        y_TP = len(np.where(y * y_hat == 1)[0])
        y_TN = len(np.where(y_hat - y == 0)[0])
        total_pixels = y.shape[0] * y.shape[1]
        
        """print('y_TP = ' + str(y_TP))
        print('y_TN = ' + str(y_TN))"""
        
        accuracy = ((y_TP + y_TN) / total_pixels)
        total_accuracy += accuracy
        processed += 1
        print('Accuracy: ' + str(accuracy * 100) + ' %.')
    except:
        print('Predicted label not in labels.')
    
    if processed == 2:
        break

print('Total Accuracy: ' + str((total_accuracy / processed) * 100) + ' %.')
```
